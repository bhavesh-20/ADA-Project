# Multivariate Inferential statistics involving two populations

Null Hypothesis: Diffenece between means of both populations is 0.

Assumpotion : Both populations have same covarinace matrix.

```python
from scipy import stats

def Spooled(df1,df2):
    n1 = len(df1)
    n2 = len(df2)
    S1 = df1.cov()
    S2 = df2.cov()
    Spool = ((n1-1)*S1 + (n2-1)*S2)/(n1+n2-2)
    return Spool

def CriticalValue(df1,df2,confidence):
    alpha = 1 - confidence
    n1 = len(df1)
    n2 = len(df2)
    p = len(df1.columns)
    fval = ((n1+n2-2)*p)/(n1+n2-p-1)*stats.f.ppf(alpha, p, n1+n2-p-1)
    return fval

def T2(df1,df2):
    n1 = len(df1)
    n2 = len(df2)
    Spool = Spooled(df1,df2)
    Spool_inv = np.linalg.inv((1/n1 + 1/n2)*Spool)
    z = df2.mean() - df1.mean()
    T2val = z.T @ Spool_inv @ z 
    return T2val

def inference(df1,df2,confidence):
    t2 = T2(df1,df2)
    f = CriticalValue(df1,df2,confidence)
    print(f"Hotelling Tsquare value - {t2}")
    print(f"Critical value - {f}")
    if t2 > f:
        return "Reject H0"
    else:
        return "Fail To Reject H0"
    
print(f"Inference when fire=1 : {inference(df1,df2,0.90)}\n")
print(f"Inference when fire=0 : {inference(df3,df4,0.90)}")
```
